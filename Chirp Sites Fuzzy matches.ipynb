```python
#Import libraries
import pandas as pd
from difflib import get_close_matches

# Load your CSVs
df = pd.read_csv(r"C:\Users\Isaiah.Bennett\Downloads\random helpdesk\chirp_sites_core.csv", encoding="ISO-8859-1", low_memory=False)

# Clean and normalize the columns
Site_Name = df["Site_Name"].dropna()
acr_names = df["acr_names"].dropna()

# Create list of matches
matches = []

for name in Site_Name:
    close = get_close_matches(name, acr_names, n=1, cutoff=0.6)
    matches.append({
        "Sitename": name,
        "acr_names": close[0] if close else None
    })

# Turn into a DataFrame
matches_df = pd.DataFrame(matches)

# Save to CSV for review
matches_df.to_csv("fuzzy_matches_outputraw.csv", index=False)

print("✅ Fuzzy match complete. Output saved to 'fuzzy_matches_output.csv'.")
```
