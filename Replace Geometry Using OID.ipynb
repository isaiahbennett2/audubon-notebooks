```python
import arcpy

source_fc = "NYS Tax Parcels Public\\NYS_Tax_Parcels_Public" #Replace with unique parcel layer
target_fc = "Spaces Places Points Polygons Guid\\spaces_places_polygons_new_relate"


# Provide the ObjectIDs explicitly
source_oid = 763360 # Replace with your actual source ObjectID
target_oid = 335  # Replace with your actual target ObjectID

# --- Get geometry from source feature ---
source_geom = None
source_where = f"OBJECTID = {source_oid}"
with arcpy.da.SearchCursor(source_fc, ["SHAPE@"], source_where) as cursor:
    for row in cursor:
        source_geom = row[0]

if source_geom is None:
    raise Exception("Could not find source geometry with that ObjectID.")

# --- Update geometry in target feature ---
target_where = f"OBJECTID = {target_oid}"
print(target_where)
updated = False
with arcpy.da.UpdateCursor(target_fc, ["SHAPE@"], target_where) as cursor:
    try:
        for row in cursor:
            row[0] = source_geom
            cursor.updateRow(row)
            updated = True
    except Exception as e:
        print(e)

if updated:
    print(f"SUCCESS: Geometry from OBJECTID {source_oid} successfully copied to OBJECTID {target_oid}.")
else:
    print(f"WARNING: Target OBJECTID {target_oid} not found or could not be updated.")
```
